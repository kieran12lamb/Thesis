```python
mutation_type = 'Amino'
# mutation_type = 'Nucleotide'

# normalised = 'Reference'
normalised = 'Normalised'

#Set Co/Me
co_or_me = 'CO'
# co_or_me = 'ME'
```

GET DATA IN CORRECT FORMAT FOR SANKEY

```python
# num_cols = len([col for col in test_sk .columns if 'source' in col ])

# for i in range(num_cols):
#     test_sk.iloc[:,num_cols-i-1] = 's|'+test_sk.iloc[:,list(range(0,num_cols-i))].agg("|".join, axis=1)
#     test_sk.iloc[:,num_cols+i] = 't|'+test_sk.iloc[:,list(range(num_cols+i,num_cols+num_cols))[::-1]].agg("|".join, axis=1)
```

```python
# Splitting initial source and target mutation names by pipe
df_sankey = df4
if mutation_type == 'Nucleotide':
    df_sankey[['source_mut', 'source_alt_aa', 'source_aa_pos', 'source_ref_aa', 'source_window', 'source_domain', 'source_protein', 'source_orf']] = df_sankey['source'].apply(lambda x: pd.Series(str(x).split('|')))
    df_sankey[['target_mut', 'target_alt_aa', 'target_aa_pos', 'target_ref_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf']] = df_sankey["target"].apply(lambda x: pd.Series(str(x).split("|")))
elif mutation_type == 'Amino':
    df_sankey[['source_alt_aa', 'source_aa_pos', 'source_ref_aa', 'source_window', 'source_domain', 'source_protein', 'source_orf']] = df_sankey['source'].apply(lambda x: pd.Series(str(x).split('|')))
    df_sankey[['target_alt_aa', 'target_aa_pos', 'target_ref_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf']] = df_sankey["target"].apply(lambda x: pd.Series(str(x).split("|")))
feature_dict = {}


for feature in ['nucleotide','amino_acid','window','domain']:
    if feature == 'nucleotide':
        if mutation_type =='Amino':
            continue
        if co_or_me == 'CO':
            columns = ['source_orf', 'source_protein', 'source_domain', 'source_window', 'source_ref_aa', 'source_aa_pos', 'source_alt_aa', 'source_mut', 'target_mut', 'target_ref_aa', 'target_aa_pos', 'target_alt_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf', 'value','frequency','mean','pv (fisher)','pv (ws)']
        else:
           columns = ['source_orf', 'source_protein', 'source_domain', 'source_window', 'source_ref_aa', 'source_aa_pos', 'source_alt_aa', 'source_mut', 'target_mut', 'target_ref_aa', 'target_aa_pos', 'target_alt_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf', 'value','pv (ws)'] 
        df_sankey = df_sankey.loc[:,columns]
    elif feature == 'amino_acid':
        if co_or_me == 'CO':
            columns = ['source_orf', 'source_protein', 'source_domain', 'source_window', 'source_ref_aa', 'source_aa_pos', 'source_alt_aa', 'target_ref_aa', 'target_aa_pos', 'target_alt_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf', 'value','frequency','mean','pv (fisher)','pv (ws)' ]
        else:
            columns = ['source_orf', 'source_protein', 'source_domain', 'source_window', 'source_ref_aa', 'source_aa_pos', 'source_alt_aa', 'target_ref_aa', 'target_aa_pos', 'target_alt_aa', 'target_window', 'target_domain', 'target_protein', 'target_orf', 'value','pv (ws)']
        df_sankey = df_sankey.loc[:,columns]
    elif feature == 'window':
        if co_or_me == 'CO':
            columns =['source_orf', 'source_protein', 'source_domain', 'source_window','target_window', 'target_domain', 'target_protein', 'target_orf', 'value','frequency','mean','pv (fisher)','pv (ws)' ]
        else:
            columns =['source_orf', 'source_protein', 'source_domain', 'source_window','target_window', 'target_domain', 'target_protein', 'target_orf', 'value','pv (ws)' ]
        df_sankey = df_sankey.loc[:,columns]
    elif feature == 'domain':
        if co_or_me == 'CO':
            columns = ['source_orf', 'source_protein', 'source_domain','target_domain', 'target_protein', 'target_orf', 'value','frequency','mean','pv (fisher)','pv (ws)' ]
        else:
            columns = ['source_orf', 'source_protein', 'source_domain','target_domain', 'target_protein', 'target_orf', 'value','pv (ws)']
        df_sankey = df_sankey.loc[:,columns]        

    # take source orf, source nucleotide, target nucleotide and target orf columns 
    new_df = df_sankey.loc[:, columns]
    
    if mutation_type == 'Nucleotide':
        if feature == 'nucleotide' or feature == 'amino_acid':
            new_df['source_ref_aa'] = new_df['source_ref_aa']+new_df['source_aa_pos']
            # +new_df['source_alt_aa']
            new_df['target_ref_aa'] = new_df['target_ref_aa']+new_df['target_aa_pos']
            # +new_df['target_alt_aa']

            new_df.rename({'source_ref_aa':'source_aa','target_ref_aa':'target_aa'},axis=1,inplace=True)
            new_df.drop(['source_aa_pos','source_alt_aa','target_alt_aa', 'target_aa_pos', ],axis=1,inplace=True)
    
    # add an 's' to the source columns and a 't' to the target columns to make them unique and avoid circular links problem 
    num_cols = len([col for col in new_df.columns if 'source' in col ])
    for i in range(num_cols):
        new_df.iloc[:,num_cols-i-1] = 's|'+new_df .iloc[:,list(range(0,num_cols-i))].agg("|".join, axis=1)
        new_df.iloc[:,num_cols+i] = 't|'+new_df.iloc[:,list(range(num_cols+i,num_cols+num_cols))].agg("|".join, axis=1)
        
    final_sankey_list = []
    source_vals = []
    num_cols = len([col for col in new_df.columns if 'source' in col ])
    for i in range(0,(num_cols*2)-1):
        index,source_values,target_values = i,new_df.iloc[:,i],new_df.iloc[:,i+1]
        source_vals.append(source_values)
        if co_or_me == 'CO':
            df = pd.DataFrame({'source':source_values,
                               'target':target_values,
                               'frequency':new_df.frequency,
                               'mean_distance':new_df['mean'],
                               'pv (fisher)':new_df['pv (fisher)'],
                               'pv (ws)':new_df['pv (ws)']})
            final_sankey_list.append(df)
            
        elif co_or_me == 'ME':
            df = pd.DataFrame({'source':source_values,
                               'target':target_values,
                                'pv (ws)':new_df['pv (ws)']})
            final_sankey_list.append(df)
    final_sankey = pd.concat(final_sankey_list,axis=0)   
    final_sankey['value'] = 1
    
    final_sankey.drop_duplicates().to_csv('Output/Generated_Sequences/'+normalised+'/Sankey/'+co_or_me+'/'+mutation_type+'/'+feature+'.csv')
```
